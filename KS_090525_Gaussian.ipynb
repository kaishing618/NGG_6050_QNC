1) find a paper that shows data thought to come from one of these distributions, and 

2) write code to simulate data that (roughly) match the distribution shown in the paper. 

Neumann et al., 2024, Science Translational Medicine

Brainwide silencing of prion protein by AAV-mediated delivery of an engineered compact epigenetic editor

Figure 4H



```python
# Gaussian distributed dataset for log10 mean Prnp intensity
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution (adjust mean and std for best match to your data)
mean_intensity = 3.1
std_intensity = 0.2
n_neurons = 10000
gaussian_data = np.random.normal(loc=mean_intensity, scale=std_intensity, size=n_neurons)

# Plot histogram with outlined bins and grey color
bins = np.linspace(2.4, 3.8, 50)
counts, edges = np.histogram(gaussian_data, bins=bins)
xaxis = (edges[1:] + edges[:-1]) / 2

plt.figure(figsize=(8,4))
plt.bar(xaxis, counts, width=(edges[1]-edges[0]), color='grey', alpha=0.8, edgecolor='black', linewidth=1)
plt.xlabel('log10 mean Prnp intensity')
plt.ylabel('Count (neurons)')
plt.title('Gaussian Distribution of log10 Mean Prnp intensity')
plt.xlim(2.4, 3.8)
plt.ylim(0, 650)
plt.show()
```

```python
# Skewed dataset for log10 mean Prnp intensity
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Parameters for the skewed distribution (adjust a, loc, scale for best match)
a = 8  # positive values = right skew
mean_intensity = 2.8
std_intensity = 0.19
n_neurons = 20000
skewed_data = skewnorm.rvs(a, loc=mean_intensity, scale=std_intensity, size=n_neurons)

# Plot histogram with outlined bins and grey color
bins = np.linspace(2.4, 3.8, 60)
counts, edges = np.histogram(skewed_data, bins=bins)
xaxis = (edges[1:] + edges[:-1]) / 2

plt.figure(figsize=(8,4))
plt.bar(xaxis, counts, width=(edges[1]-edges[0]), color='#003A6B', alpha=0.8, edgecolor='black', linewidth=1)
plt.xlabel('log10 mean Prnp intensity')
plt.ylabel('Count (neurons)')
plt.title('Skewed Distribution of log10 Mean Prnp intensity')
plt.xlim(2.4, 3.8)
plt.ylim(0, 2000)
plt.show()
```
